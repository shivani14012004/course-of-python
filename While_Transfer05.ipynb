```python
#Continue:will help to continue the loop and skip current condition or item
#Eg:-
candidate=[
    {'name':'Gaytri','status':'present'},
    {'name':'aarya','status':'absent'},
    {'name':'shruti','status':'present'},
    {'name':'neha','status':'absent'},
 ]
for student in candidate:
  if student['status'] == 'absent':
    print(student['name'],'is absent')
    continue
  else:
    print('Give an assignment to:',student['name'])

```

```python
marks = [85,0,90,76,0,70,68,-1,88,96]
for mark in marks:
  if mark ==-1:
    print('Invalid marks found,Stop checking')
    break

  if mark == 0:
    print('Absent student,skip this student')
    continue

  if mark<40:
    pass

  print('valid marks:',mark)
```

#**List Comprehension
#-to create a new list in simple one liner solution
#-it uses the tempeary operator to check if else option

```python
#Syntax:[operation for val in sequence/iteration]
```

```python
d=[2,3,4,5,6,7]
#create a new list with squared numbers
[num *num for num in d]
```

```python
nms=['prasad','shivani','aarya','shruti']
[nm.title() for nm in nms]
```

```python
[(nm.title(),len(nm)) for nm in nms] #(name,len(name))
```

```python
[nm.upper() for nm in nms]
```

```python
[(nm.capitalize(),len(nm)) for nm in nms]
```

```python
#Interview Q1)Coditional list comprehension
#fetch those names whoselen is>6
#Syntax:[operation for val in sequence/iteration if<condition>]
#first element fetched-->check condition -->perform operation
[nm.title() for nm in nms if len(nm)>6]
```
