```python
#Eg:-Direct condition
[i+100 for i in range (1,11)]
```

```python
#1)[operation_var for var in sequence]
#2)with condition if<condition>
#3)[opertion for var in sequence if<condition>]
```

```python
#Interview Q1)Need a list of Even numbers?
#eg:-if condition
[num for num in range(1,21)if num%2==0]
```

```python
#3)Ternary operator
#syntax:-
#express_1 if <condition>else expression_2
#eg:-
gender=['Male','Female','Female','Male']
#convert 'Male' to 0 and Female to 1
[0 if val=='Male' else 1 for val in gender]
```

```python
salary=[24000,12500,45000,65000,11000,10000]
#PS:if sal>25000 then 10% bonus else 20%
[sal+sal*.1 if sal>25000 else sal+sal*.2 for sal in salary]
```

```python
#Explore Dictionary comprehension
#Create a dict {num:num^2}(1-10)
#Eg:-
[{val:val*val}for val in range(1,11)]
```

```python
#Nested List Comprehension
#new_list=Expression for item in inner_iterable]for item in outer_iterable]
#Eg:-
[[val2 for val2 in range(2)]for val in range(1,4)]
```

```python
for val in range(2):
  print(val)
```

```python
[[val for val in range(1,4)]for val in range(2)]
```

```python
[[val for val in range(1,4)]for val in range(1,4)]
```

```python
matrix=[[1,2,3],[4,5,6],[7,8,9]]
odds=[num for row in matrix for num in row if num%2!=0]
print(odds)
```

```python
[num for num in  matrix]
```

```python
[row for row in matrix]
```

```python
[num for row in  matrix for num in row]
```

```python
[num for row in  matrix for num in row if num%2==0]
```
