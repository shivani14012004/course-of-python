#Numpy Slicing and Indexing

```python
x=np.arange(24).reshape((6,4))
x
```

```python
#Using index we can access one row at a time
x[3]
```

```python
#row with element index
#1)
x[3][3]
```

```python
#2)
x[3,3]
```

```python
#Examples
x[2]
```

```python
x[2,3]
```

```python
x[2][3]
```

```python
#Numpy array is mutable in nature
x[3][3]
```

```python
#replace 15 by 50
x[3,3]=50
x
```

```python
#access more rows
x[:]#[start_index : stop_index]
```

```python
x[1:4]
```

```python
x[3:] #+ve
```

```python
x[-3:] #-ve
```

```python
x[::-3]
```

```python
x[::-2]
```

```python
x[::-4]
```

```python
#column operation
x[-2:,:] #[row,column]
```

```python
x[-2:,-2:]
```
