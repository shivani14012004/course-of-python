#*Function with arguments

```python
def addition(x,y): #declaration
  print('Addition:',x+y)
  print('Division:',x/y)
  print('Multiplication:',x*y)
  print('Substraction:',x-y)

addition(100,200) #calling

```

```python
def addition(y,x): #declaration
  print('Addition:',x+y)
  print('Division:',x/y)
  print('Multiplication:',x*y)
  print('Substraction:',x-y)

addition(100,200)
```

#Types of arguments

```python
#1)positional args
#2)Keyword args
#3)Variable length args
#4)Default args
```

```python
#1).Positional Argument
#these are args. in which sequence order matters
#Eg:-
def test(x,y,z):
  print(x,y,z)
test(100,200,300)
test(200,300,400)
```

```python
def test(z,x,y):
  print(z,x,y)
test(100,200,300)
test(200,300,400)
```

```python
#Eg 2:-
def info(name,age,sal):
  print('your name is:',name,'Your age is:',age,'& Your salary is:',sal)
info('Shivani',21,60000)#position argument
```

```python
info(60000,'shivani',21) #When user doesnt follow order then output is not as per the expectation.
```

```python
#Keywords args:-
#to solve above problem we have keyword args.
#which doesnt require to follow an order
#bcz it directly refers to key=value format

info(sal=60000,name='Shivani',age=21)
```

```python
info('Shivani',21)#we must supply 3 values to 3 args
```

```python
info('Shivani',21,None)# when we not want to send
```

```python
info(name='Shivani')#in keyword also we need to supply the values= to number of args
```

```python
#Default args:-
#to solve above problem we have default arguments
#using default args we set the static values in declaration
#so when user is not supplying the values from calling the sytesm will take these default value

def info(name='Guest',age=23,sal=25000):
  print('Your name is:',name,'Your age is:',age,'your sal is:',sal)
info()
```

```python
#4)Variable length arguments
#1)variable length position args

#2)Variable length keyword args
```

```python
#1)variable length position args-->*args
#when user doesnt have a strict requirement of string the values to the arguments
#and values varies from one calling to another then use var.len.positional args.
#we need to prefix the argument
def test(*n):
  print(n)
test(10)
```

```python
test(10,20,30)

test('python',34,23432432,'c++')
```

```python
test()
```

```python
#2).variable length keyword args-->**kwargs
def test(**x):
  print(x)
test(num1=10,num2=20,num=30)
```

```python
test(name='Shivani',age=23,salary=4500)
```

```python
#when we want to combine both we can do so
#but remember the rule
#first we need positional n then keyword

def sample(*args,**kwargs):
  print(args,kwargs)

sample(10,20,30,num1=10,num2=20,num3=300)
```
