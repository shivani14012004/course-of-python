```python
#Create a simple lambda function that adds 10 to a given number.
add=lambda x:x+10
print(add(30))
```

```python
#Create a lambda function to multiply two numbers.
multiply=lambda x,y:x*y
print(multiply(4,3))
```

```python
# Create a lambda function that returns the square of a given number.
sqr=lambda x:x**2
print(sqr(3))
```

```python
#Create a lambda function to check whether a number is even or odd.
num=lambda x:'even' if x%2==0 else'odd'
print(num(3))
print(num(2))
```

```python
#Use lambda with map() to square every number in a list.
num=[2,3,4,5,6]
list(map(lambda x:x*x,num))
```

```python
#Use lambda with filter() to get only even numbers from a list.
num=[2,3,4,5,6,7]
list(filter(lambda x:x%2==0,num))
```

```python
# Sort a list of tuples based on the second value using lambda.
students=[('amit',89),('atharv',88),('shivraj',84),('shivani',78)]
sorted(students,key=lambda x:x[1])
```

```python
# Sort employee records based on salary using lambda.
employee=[
    {'name':'jyoti','salary':20000},
    {'name':'sakshi','salary':3000},
    {'name':'shrutika','salary':23000}
]
sorted(employee,key=lambda x:x['salary'])
```

```python
#Create a lambda function that returns the maximum of two numbers.
max=lambda a,b:a if a>b else b
print(max(23,45))
```

```python
# Create a function that returns a lambda function for multiplying a number by a given value.
def multiplier(n):
  return lambda x:x*n
double =multiplier(2)
triple =multiplier(3)
print(double(10))
print(triple(10))
```

```python
#Use lambda with reduce() to find the product of all numbers in a list.
from functools import reduce
numbers =[1,2,3,4,5]
reduce(lambda a,b:a*b,numbers)
```

```python
#Problem Statement: Create lambda functions to find the square and cube of a number.
sqr=lambda num:num*num
print(sqr(2))

cube=lambda num:num*num*num
print(cube(2))

```

```python
# Problem Statement: Write a lambda function to check whether a number is positive, negative, or zero.

check=lambda x:'positive' if x>0 else 'negative'if x< 0 else 'zero'
print(check(2))
print(check(-1))
print(check(0))
```

```python
# Problem Statement: Use lambda and filter() to select words starting with the letter A.
words=['Apple','banna','Mango','Apple','grapes']
list(filter(lambda word: word.startswith('A'),words))
```

```python
# Problem Statement: Use lambda and map() to convert a list of strings into uppercase
words=['aarya','shivani','sakshi','shivraj']
list(map(lambda word:word.upper(),words))
```

```python
#Problem Statement: Sort student records by marks from highest to lowest using lambda.
students=[('sakshi',84,),('shruti',87),('jyoti',67)]
sorted(students,key=lambda x:x[1],reverse=True)
```

```python
#Problem Statement: Sort a list of employee names based on the length of each name.
employee=['shivani','sakshi','aarya','jyoti']
sorted(employee,key=lambda name:len(name))
```
