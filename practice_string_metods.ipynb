```python
#by using for loop and split()
text='I like java'
words=text.split()
result=''
for word in words:
  if word =='java':
    result += 'python'
  else :
    result += word +' '
print(result)
```

```python
#by using for loop and continue
text='I like java'
words = text.split()
result=''
for word in words:
  if word =='java':
    result +='python'
    continue
  result += word +' '
print(result)

```

```python
#Return Function (Using a for Loop)
def reverse_string():
  text='python'
  reverse=''

  for ch in text:
    reverse=ch + reverse

  return reverse
reverse_string()

```

```python
#Using a while Loop
def reverse_string():
  text='python'
  reverse=''
  i=len(text)-1

  while i>=0:
    reverse += text[i]
    i-= 1
  return reverse
reverse_string()
```

```python
#Using a for Loop and if-else
def reverse_string():
  text='python'
  reverse=''
  for ch in text:
    if ch:
      reverse=ch+reverse
    else:
      reverse=reverse
  return reverse
reverse_string()
```
