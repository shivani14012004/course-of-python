```python
#Square Every Number
numbers = [1,2,3,4,5]
list(map(lambda x:x*x,numbers))
```

```python
#2. Cube Every Number
numbers = [2,3,4]
list(map(lambda x:x**3,numbers))
```

```python
#Convert to Uppercase
str =['python','java','c++']
list(map(lambda x:x.upper(),str))
```

```python
#Find Length of Each Word
names=['shivani','patil','shrutila']
list(map(lambda x:len(x),names))
```

```python
#Add 10 to Every Number
numbers=[10,20,30]
list(map(lambda x:x+10,numbers))
```

```python
#Double Every Number
num=[10,20,30]
list(map(lambda x:x*2,num))
```

```python
#Multiply Two Lists
a =[1,2,3]
b =[2,3,1]
list(map(lambda x,y:x*y,a,b))
```

```python
#Capitalize First Letter
names = ['shivani','python']
list(map(lambda x:x.capitalize(),names))
```

```python
#Convert Celsius to Fahrenheit
temp=[0,20,30]
list(map(lambda c:(c*9/5)+32,temp))
```

```python
#Reverse Every String
words=['shivani','sakshi','shrutika']
list(map(lambda x:x[::-1],words))
```

```python
#map() with if-else (Even/Odd)
number=[1,2,3,4]
list(map(lambda x:'even' if x%2==0 else 'odd',number))
```

```python
#map() with if-else (Positive/Negative)
nums=[1,2,-1,-11,-22,12,23]
list(map(lambda x:'positive'if x>0 else 'negative',nums))
```
