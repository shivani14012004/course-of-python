```python
#3)Reduce
#reduce(function,iterable)--> it is used to convert collection into a single value/object

#reduce is present in functools module

# For example, reduce(lambda x, y: x+y, [1, 2, 3, 4, 5])
# calculates ((((1 + 2) + 3) + 4) + 5)

from functools import reduce

k=[2,3,4,5,6]
reduce(lambda x,y:x+y,k)#in reduce we dont need typecasting as it gives the result directly
```

```python
reduce(lambda x,y:max(x,y),[0,12,34,555,23,1])
```
