#2.Iterative statement

```python
#It contains 2 loops
#1.for loop
#2.while loop
```

```python
#1.forloop:- is an iterative loop which fetches on element at a time from the collection inside the body of
a=[10,20,30,40]

#syntax for var in collection/iterable
for val in a:
  print(val)
```

```python
name='vandana'
for i in name:
  print(i)
```

```python
a=[10,20,30,40]

#syntax for var in collection/iterable
for val in a[::-1]:
  print(val)
```

```python
#Assignment:-Take a dict and apply for loop
#simple
a={1:100,2:200}
for val in a:
  print(val)
```

```python
a={1:100,2:200}
for val in a:
  print(val,a[val])
```
