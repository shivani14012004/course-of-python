```python
try:
# print(30+20)#line without error
# print(30+'p')#line with error
# print(30/0)
# print(30+a)
  print(int('ten'))
except TypeError as msg:
  print('TypeError:',msg)
except ValueError as msg:
  print('Error occur:',msg)
```

#>try:test code
#>except:to handle exception
#>else:if no exception then come here
#>finally :irrespective of exception this block will get executed

```python
try:
# print(10/4)without exception
# print(100/0)#with exception ZerodivisionError

  print(int('ten'))#with exception ValueError
except Exception as msg:
  print(msg)
else:
  print('No exception do below tasks..')
finally:
  print('Zukega nai sala..')
```

```python
#Example of ATM operation

try:
  balance = 5000
  amount = int(input('Enter amout in Rs.'))

  if  amount > balance:
    raise ValueError('Insufficient balance..')

except ValueError as e:
  print('Error:',e)

else:
  balance = balance - amount
  print('Withdrawal succesfull....')
  print('Remaining Amount is Rs. ',balance)

finally:
  print('Thank you for banking with us Visit Again.. :-)')
```

```python
#Nested try Except
try:
  pin = int(input('Enter your 4 digit pin:'))
  try:
    balance = 5000
    amount = int(input('Enter amout in Rs.'))

    if  amount > balance:
      raise ValueError('Insufficient balance..')

  except ValueError as e:
    print('Error:',e)

  else:
    balance = balance - amount
    print('Withdrawal succesfull....')
    print('Remaining Amount is Rs. ',balance)

  finally:
    print('Thank you for banking with us Visit Again.. :-)')
except ValueError:
  print('Invalid PIN please enter 4 digit number ONLY')
```
