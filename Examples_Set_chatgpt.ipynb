```python
#1. Remove Duplicate Elements from a List
numbers = [10, 20, 10, 30, 20, 40]
unique=list(set(numbers))
print(unique)
```

```python
#2. Find Common Elements Between Two Sets
A = {10, 20, 30, 40}
B = {30, 40, 50, 60}
common = A & B
print(common)
```

```python
#3. Find Union of Two Sets
A = {1,2,3}
B = {3,4,5}
print (A|B)
```

```python
#4. Find Difference Between Two Sets
A = {1,2,3,4}
B = {3,4,5}
print(A-B)
```

```python
#6. Check if One Set is Subset of Another
A = {1,2}
B = {1,2,3,4}
print(A.issubset(B))
```

```python
#7. Check Superset
A={1,2,3,4}
B={1,2}
print(A.issuperset(B))
```

```python
#8. Remove an Element
fruits={'apple','grapes','mango'}
fruits.remove('grapes')
print(fruits)
```

```python
#9. Add Multiple Elements
A={1,2,3}
A.update([4,5,6])
print(A)
```

```python
#10. Remove Duplicate Characters from String
text='programming'
print(set(text))
```

```python
#11. Count Unique Elements
num=[1,2,2,3,4,4,5,6]
print(len(set(num)))
```

```python
#12. Are Two Lists Equal?
a=[1,2,3,4]
b=[4,3,2,1]
print(set(a)==set(b))
```

```python
#13. Find Missing Numbers
num={1,2,3,4,5,6,7,8,9}
present={1,2,4,6,7,9}
missing=num-present
print(missing)
```

```python
#14. Convert List to Set
num=[1,2,3,4,5]
s=set(num)
print(s)
```

```python
#19. Remove Random Element
a={10,20,30}
num=a.pop()
print(num)
print(a)
```

```python
#20. Set Comprehension
num=[1,2,3,3,4,4,5,6]
result={x*x for x in num}
print(result)
```
