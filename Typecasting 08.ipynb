```python
#Assignment:-Crete a simple bill of Grocery.Take 3-4 inputs and print the final bill.
#ex:-

item1=eval(input('Enter a amount of item1:'))
item2=eval(input('Enter a amount of item2:'))
item3=eval(input('Enter a amount of item3:'))
item4=eval(input('Enter a amount of item4:'))

print('Final Billing is',item1+item2+item3+item4)
```
