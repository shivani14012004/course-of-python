#Lambda Function

```python
#anonymous/nameless function
#It is a one liner function
#very important Interview Q:-syntax:-lambda parameter/s:expression
```

```python
#squaring of number
sq=lambda num:num*num
sq(8)
```

```python
#only one expression is allowed in lambda
#eg:-
sq=lambda num:num*num,num+100 #here we are adding one more expression which is not allowed
sq(8)
```

```python
#lambda function by default return the value of the expression
change=lambda name:name.upper()

result=change('shivani')
print(result)
```

```python
#what is difference between normal function and lambda function?
#when to use lambda function??
```

```python
#use of ternary operator in lambda
test = lambda num:'+ve' if num>0 else '-ve'
test(90)
```

```python
test(-90)
```

```python
#check if number is even or odd
check = lambda num:'Even' if num%2==0 else None
check(8)
```

```python
check(7)
```

```python
check = lambda num:'True' if num%2==0 else None
check(8)
```

#**Higher order Function:-It is a function who need another function as an input
#function(function)


```python
#Three types of Higher order function
#1).Map
#2).filter
#3).reduce
```

```python
#Iterable:-cllection of element -->str,list,tuple,set,dict,range,array..

#map(function,iterable)

k=[20,10,3,4,5]
#apply an operation over each element from an iterable

map(lambda num:num*num,k)
```

```python
 0x798e9d456b90
```

```python
#Rule-When we are getting an output in the form of object then we have 2 choices
#Choice1-typecast the object to see actual values
tuple(map(lambda num:num*num,k))
```

```python
#Choice 2- use for loop
out=map(lambda num:num*num,k)
for val in out:
  print(val)
```

```python
#can we use more args in lambda:Yes
test=lambda a,b,c:a+b
test(100,800,500)
```

```python
#logially we can do
test=(lambda a,b,c:max(a,b,c))
test(100,800,500)
```

```python
names=('gayatri','shivani','aarya','shruti')
#convert above names in Upper case using map
map(lambda upper_case:upper_case.upper(),names)
```

```python
list(map(lambda names:names.upper(),names))
```

```python
tuple(map(lambda names:names.upper(),names))
```

```python
#2).Filter(function,iterable)
#this is conditional function
#the value for which condition is True, will get fetched and operation will be performed if requried
# filter is not for operation

p=(12,-33,40,-8,99,-12)
#filter -ve numbers
tuple(filter(lambda num:num<0,p))#-ve numbers
```

```python
tuple(filter(lambda num:num>0,p))#+ve numbers
```
