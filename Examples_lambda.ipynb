```python
#Add Two Numbers
add=lambda a,b:a+b
print(add(10,20))
```

```python
squares=lambda num:num*num
print(squares(5))
```

```python
cube =lambda x:x**3
print(cube(2))
```

```python
max=lambda a,b: a if a>b else b
print(max(20,10))
```

```python
min=lambda a,b:a if a<b else b
print(min(20,10))
```

```python
from typing import OrderedDict
#Even and odd
check=lambda a:'Even' if a %2==0 else None
print(check(2))
```

```python
check=lambda b:'even' if b %2==0 else 'odd'
print(check(3))
```

```python
check=lambda c:'Even' if c%2==0 else'None'
print(check(3))
```

```python
check=lambda a:'positive' if a>=0 else 'Negative'
print(check(48))
```

```python
name=lambda str:len(str)
print(name('python'))
```

```python
#Last Character
name=lambda a:a[-1]
print(name('shivani'))
```

```python
#first character
name=lambda s:s[0]
print(name('shivani'))
```

```python
#10. Multiply Three Numbers
multiply=lambda a,b,c: a*b*c
print(multiply(2,3,4))
```

```python
leap=lambda year:'leap year' if year % 400==0 or (year %4 ==0 and year % 100!=0) else 'not leap year'
print(leap(2024))
```

```python
#. Check Vowel
vowels=lambda ch:'vowels' if ch.lower() in 'aeiou' else 'consonant'
print(vowels('p'))
```

```python
vowel=lambda ch:'vowel' if ch.lower() in 'aeiou'else 'consontant'
print(vowel('o'))
```

```python
#20. Check Palindrome
palindrome=lambda s:'palindrome' if s==s[::-1] else 'not palindrome'
print(palindrome('madam'))
```

```python
#21. Factorial (Using Lambda + Recursion)
fact=lambda n: 1 if n==0 else n*fact(n-1)
print(fact(5))
```

```python
#25. Find Larger String
longer=lambda a,b:a if len(a)>len(b) else b
print(longer('python','c++'))
```

```python
#Sort List of Dictionaries
Employees=[
    {'nmaes':'Sakshi','salary':30000},
    {'nmaes':'Shruti','salary':50000},
    {'nmaes':'Shivarth','salary':40000},
]
result=sorted(Employees,key=lambda x:x['salary'])
print(result)
```
