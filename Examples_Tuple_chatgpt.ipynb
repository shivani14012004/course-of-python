```python
#1. Create a Tuple
num=(10,20,30,40)
print(num)
```

```python
#2. Access Tuple Elements
fruits=('Apple','Mango','Grapes','Banana')
print(fruits[1])
```

```python
#3. Find Length of Tuple
num=(10,20,30,40)
print(len(num))
```

```python
#4. Count Occurrences of an Element
num=(1,2,3,4,2,2)
print(num.count(2))
```

```python
#5. Find Index of an Element
num=(10,20,30,40,50)
print(num.index(30))
```

```python
#6. Check if an Item Exists
colours=('Blue','Red','Yellow')
if 'Blue' in colours:
  print('fond')
else:
  print('not found')
```

```python
#7. Concatenate Two Tuples
a=(1,2)
b=(3,4)
result = a + b
print(result)
```

```python
#8. Repeat Tuple
t=("Python",)
print(t * 3)
```

```python
#9. Slice a Tuple
num=(10,20,30,40,50)
print(num[1:4])
```

```python
#10. Reverse a Tuple
num=(1,2,3,4,5)
num[::-1]
```

```python
num=(1,2,3,4,5)
reverse=[]
for i in (range(num[-1])):
  reverse.append(i)
print(reverse)
```

```python
#11. Find Maximum Element
num=(10,20,30,50)
print(max(num))
```

```python
##11. Find Minimum Element
num=(10,20,30,50)
print(min(num))
```

```python
#Find Sum of Tuple
num=(1,2,3,4,5)
print(sum(num))
```

```python
#14. Convert List to Tuple
num=[1,2,3,4]
result=tuple(num)
print(result)
```

```python
#15. Convert Tuple to List
num=(1,2,3,4)
result=list(num)
print(result)
```

```python
#16. Swap Two Tuples
t1 = (1,2)
t2 = (3,4)
t1,t2=t2,t1
print(t1)
print(t2)
```

```python
#17. Unpack Tuple
student=('shivani',23,'python')
name,age,course=student
print(name)
print(age)
print(course)
```

```python
#18. Nested Tuple
#For example:
# students[0][0] → "Ram"
# students[0][1] → 80
# students[1][0] → "Shyam"
# students[1][1] → 90
# students[2][0] → "Mohan"
# students[2][1] → 85

student=(('shivani',80),('shruti',70),('manu',87))
print(student[1][0])
print(student[1][1])
```

```python
#19. Loop Through Tuple
num=(10,20,30,40)
for i in num:
  print(i)
```

```python
#20. Find Common Elements
t1 = (1,2,3,4)
t2 = (3,4,5,6)
common = ()
for i in t1:
  if  i in t2:
    common +=(i,)
print(common)
```

```python
#21. Remove Duplicates
num=(1,2,3,3,4,4,5,5)
result=tuple(set(num))
print(result)
```

```python
#22. Sort a Tuple

numbers = (5,2,8,1)
result=tuple(sorted(numbers))
print(result)
```

```python
#23. Find Second Largest Element
num=(10,20,40,50,30,60)
result=tuple(sorted(set(num)))
print(result[-2])
```

```python
#24. Count Even and Odd Numbers
num=(1,2,3,4,5,6)
even = 0
odd = 0
for i in num:
  if i %2 == 0:
    even +=1
  else:
    odd +=1
print(even)
print(odd)
```

```python
#25. Find the Largest String
num=(10,20,40,30,45,60)
result=tuple(sorted(set(num)))
print(result[-1])
```

```python
words = ("PHP","Exercises","Backend","Python")
longest = words[0]
for word in words:
  if len(word) > len(longest):
    longest = word
print(longest)
```
