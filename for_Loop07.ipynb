```python
#Purpose to use for loop?
#it is used to fetch the elements from collection ONE by ONE
#Eg:-
nms=['shivani','aarya','sakshi','sanu']
for n in nms:
  print(n)
```

```python
nms=['shivani','aarya','sakshi','sanu']
for n in nms:
  print(n.upper())
```

```python
nms=['shivani','aarya','sakshi','sanu']
for n in nms:
  print(n.capitalize())
```

```python
for n in nms:
  if n.endswith('i'):  #this will show paticular world which endswith 'i'.
    print(n.upper())
```

```python
for n in nms:
  if n.startswith('a'):
     print(n.upper())
```
