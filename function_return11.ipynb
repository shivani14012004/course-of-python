```python
#purchase and Billing example using return
items={'rice':56,'Chana':30,'wheat':60,'parl_g':10,'milk':35}
bill=[]
def purchase(*args):
  for item in args:
    #print(item,items[item])
    bill.append(items[item])
  return bill

def check(total):
  return sum(total)

#calling of purchase
total=purchase('wheat','milk')
print('item amounts:',total)

#Calling of check
ans = check(total)
print('Final bill amount is Rs:',ans)
```
